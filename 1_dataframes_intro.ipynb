# Drop rows with null values with dropna() method

```python
nba.tail(3)      # we have null row at the end
```

```python
nba.drop("Weightin Kilograms", axis=1, inplace=True)  # prev added extra col -- removing
```

```python
nba.drop("Salary in Millions", axis=1, inplace=True)
```

```python
nba.dropna()  # It removes all the rows with any null values.
```

```python
# check the parameters: how='any' <-- default value -- any null values drop the entire axis; default axis=0
```

```python
# how = 'all'  <-- means removes the axis if all the row values is null
nba.dropna(how = 'all', inplace = True)   # now we have no full null rows
# it gives us a brand new df also it has inplace parameter to replace permanently original df
```

```python
# also we can do for the column just we need to change the axis parameter from default value 0 to 1 or "columns"
# here also same we can mention the 'how' parameter from defautl 'any' to 'all' value

nba.dropna(axis = 'columns')   # salary and college omitted as they have any NaN values -- use how: all to remove if all the column value has null value to drop
```

```python
nba.head(3)
```

```python
# now if we don't want to drop based on the entire df has NaN or not but we can do it by just checking specific columns has a NaN value or not and based on specific column we can drop the rows
# lets say we want to drop the corresponding row, if any value has NaN value in the salary column
# 'subset' parameter where we can specify the column in an array

nba.dropna(subset = ['Salary'])
```

# astype() method for changing the data type

```python
nba["Salary"] = nba["Salary"].astype("int")      #converting float to size for salary column
# .astype doesn't have inplace parameters so reassign to the same columns is the solution.
nba.head(3)
```

```python
# small number of possible values in a column -- 'category' type can reduce it
nba["Position"].nunique()   # only 5 unique value so it's ideal to category type
nba["Position"] = nba["Position"].astype("category")   # we are not going to observe any frontend change but it does reduce the memory size
```

```python
nba.info()   # just changing two dtypes we we decreased 4 KB  -- great
```

## Sort a Data frame with the sort_values() method, Part 2 -- Sort by multiple value

```python
# It'll first look at Team then Name
nba.sort_values(["Team", "Name"])
```

```python
# sort via descending
nba.sort_values(["Team", "Name"], ascending=False)
```

```python
# sort by ascending sequence order -- first team as ascending order -- Last Name as descending order
nba.sort_values(["Team", "Name"], ascending=[True, False])
# can use inplace parameter to permanently replaced.
```

# Rank values with the .rank() method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary"].rank()    # lowest to highest
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending= False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending=False)   # make sense?
```
