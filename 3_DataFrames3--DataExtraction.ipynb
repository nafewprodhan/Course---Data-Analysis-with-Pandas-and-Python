## Use the set_index() and reset_index() methods to define a new dataframe index

```python
bond.set_index(keys="Film", inplace=True)
```

```python
bond.head()
```

```python
# what if we want to get back the numeric index back
bond.reset_index()
# index became the df first column
# drop=False --> don't drop the column which was index
bond.reset_index(drop=False)
# but what if we want to drop the prev index 
bond.reset_index(drop=True)
# but we don't here
bond.reset_index()
```

```python
bond.reset_index(inplace=True)
bond.set_index(keys="Film", inplace=True)
bond.set_index(keys="Year")  # If we do that it'll by default  remove the Film index and also Film index will not added to the coulumn
# In this case we can combine reset_index with set_index
bond.reset_index(inplace=True)
bond.set_index("Year", inplace=True)  # Now Film index move to the df column, not removing it
bond.head()
```

# Retrieve rows by index label with .loc[] Accessor

```python
bond.reset_index(inplace=True)
bond.set_index("Film", inplace=True)
bond.sort_index(inplace=True)
bond.head(3)
```

```python
 # .loc[] accessor accessing rows by its index label
bond.loc["Goldfinger"]      # return a series where column name as a index and value is the corresponding row values
    
```

```python
bond.loc["GoldenEye"]
# if index label doesn't exist it'll return a key error
```

```python
# can return a df -- if multiple rows match
bond.loc["Casino Royale"]
```

```python
# All python Slicing option applicable
bond.loc["Diamonds Are Forever":"From Russia with Love"]  # also include the last point
# slicing with step
bond.loc["Diamonds Are Forever":"From Russia with Love":2]
```

```python
bond.loc["GOldenEye":]
bond.loc[:"GOldenEye"]
```

```python
# targetted list
bond.loc[["Die Another Day", "Octopussy"]]  # follow the order as here it given
# always remember that if any value is not matched make the whole expression as am error
# to check
"Gold Bond" in bond.index
```

## Retrieve rows by index position with (numeric) iloc[] Accessor

```python
# same as loc[] but it's with number
bond.iloc[0]
```

```python
bond.iloc[15]
```

```python
bond.iloc[[15,20]]
```

```python
# same as lo[] if any index doesn't exist will return an error
```

```python
bond.iloc[4:8]
bond.iloc[20:]
bond.iloc[:4]
```

## Second Arguments to loc and iloc accessors

```python
bond.head(3)
```

```python
# By the second argument we can mention that which column we want to extract from the row, rather whole mentioned index column

bond.loc["Moonraker", "Actor"]   #just got the values
```

```python
bond.loc["Moonraker", "Director"]
```

```python
# what if we want to extract multiple column of the corresponding row
bond.loc["Moonraker", ["Director", "Box Office"]]  # result in a series
```

```python
# multiple index extraction
bond.loc[["Moonraker", "A View to a Kill"], ["Director", "Box Office"]]
```

```python
# All slicing options too -- pretty handy
bond.loc["Moonraker": "Thunderball", "Director": "Budget"]
```

```python
bond.iloc[14, 2]  # 15th row and 3rd column
```

```python
bond.iloc[14, 2:5]
```

```python
bond.iloc[[14,18,23], [2,1,4]]
```

## rename index labels or columns in a DataFrame

```python
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                     "The World Is Not Enough": "Best Bond Movie Ever"})
```

```python
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = 0)
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = "rows")
bond.rename(mapper = {"GoldenEye": "Golden Eye",
                     "The World Is Not Enough": "Best Bond Movie Ever"}, axis = "index")
bond.rename(index = {"GoldenEye": "Golden Eye",
                     "The World Is Not Enough": "Best Bond Movie Ever"})
```

```python
# rename column
bond.head(1)
```

```python
bond.rename(mapper = {"Year": "Release Date",
                     "Box Office": "Revenue"}, axis=1)
bond.rename(mapper = {"Year": "Release Date",
                     "Box Office": "Revenue"}, axis="columns")
bond.rename(columns = {"Year": "Release Date",
                     "Box Office": "Revenue"})
# to make permanent called inplace=True param
```
